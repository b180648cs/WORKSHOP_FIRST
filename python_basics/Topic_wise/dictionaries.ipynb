A Dictionary is a collection which is unordered, changeable and indexed. No duplicate members.

Create dict

```python
person = {
    'first_name': 'John',
    'last_name': 'Doe',
    'age': 30
}
```

Use constructor<br>
person2 = dict(first_name='Sara', last_name='Williams')

Get value

```python
print(person['first_name'])
print(person.get('last_name'))
```

Add key/value

```python
person['phone'] = '555-555-5555'
```

Get dict keys

```python
print(person.keys())
```

Get dict items

```python
print(person.items())
```

Copy dict

```python
person2 = person.copy()
person2['city'] = 'Boston'
```

Remove item

```python
del(person['age'])
person.pop('phone')
```

Clear

```python
person.clear()
```

Get length

```python
print(len(person2))
```

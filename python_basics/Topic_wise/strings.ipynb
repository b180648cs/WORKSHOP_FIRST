Strings in python are surrounded by either single or double quotation marks. Let's look at string formatting and some string methods

```python
name = 'Brad'
age = 37
```

Concatenate

```python
print('Hello, my name is ' + name + ' and I am ' + str(age))
```

String Formatting

Arguments by position

```python
print('My name is {name} and I am {age}'.format(name=name, age=age))
```

F-Strings (3.6+)

```python
print(f'Hello, my name is {name} and I am {age}')
```

String Methods

```python
s = 'helloworld'
```

Capitalize string

```python
print(s.capitalize())
```

Make all uppercase

```python
print(s.upper())
```

Make all lower

```python
print(s.lower())
```

Swap case

```python
print(s.swapcase())
```

Get length

```python
print(len(s))
```

Replace

```python
print(s.replace('world', 'everyone'))
```

Count

```python
sub = 'h'
print(s.count(sub))
```

Starts with

```python
print(s.startswith('hello'))
```

Ends with

```python
print(s.endswith('d'))
```

Split into a list

```python
print(s.split())
```

Find position

```python
print(s.find('r'))
```

Is all alphanumeric

```python
print(s.isalnum())
```

Is all alphabetic

```python
print(s.isalpha())
```

Is all numeric

```python
print(s.isnumeric())
```

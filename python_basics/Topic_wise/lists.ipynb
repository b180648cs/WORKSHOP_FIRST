A List is a collection which is ordered and changeable. Allows duplicate members.

Create list

```python
numbers = [1, 2, 3, 4, 5]
fruits = ['Apples', 'Oranges', 'Grapes', 'Pears']
```

Get a value

```python
print(fruits[1])
```

Get the last value

```python
print(fruits[-1])
```

Get length

```python
print(len(fruits))
```

Append to list

```python
fruits.append('Mangos')
```

Remove from list

```python
fruits.remove('Grapes')
```

Insert into position

```python
fruits.insert(2, 'Strawberries')
```

Change value

```python
fruits[0] = 'Blueberries'
```

Remove with pop

```python
fruits.pop(2)
```

Reverse list

```python
fruits.reverse()
```

Sort list

```python
fruits.sort()
```

Reverse sort

```python
fruits.sort(reverse=True)
```
